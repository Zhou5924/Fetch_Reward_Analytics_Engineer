```python
import pandas as pd 
from pandas.io.json import json_normalize
```

# Load Receipt Dataset

```python
# read json data file as pandas DataFrame
Receipt = pd.read_json('Data/receipts.json', lines=True)
Receipt.head()
```

```python
# basic information
Receipt.info()
```

# Data Processing

```python
# normalize the semi-structured data into a flat table
Receipt['_id'] = pd.json_normalize(Receipt['_id'])
Receipt['createDate'] = pd.json_normalize(Receipt['createDate'])
Receipt['dateScanned'] = pd.json_normalize(Receipt['dateScanned'])
Receipt['finishedDate'] = pd.json_normalize(Receipt[Receipt['finishedDate'].notnull()]['finishedDate'])
Receipt['modifyDate'] = pd.json_normalize(Receipt['modifyDate'])
Receipt['pointsAwardedDate'] = pd.json_normalize(Receipt[Receipt['pointsAwardedDate'].notnull()]['pointsAwardedDate'])
Receipt['purchaseDate'] = pd.json_normalize(Receipt[Receipt['purchaseDate'].notnull()]['purchaseDate'])
Receipt.head()
```

```python
# make rewardsReceiptItemlist into a separate table
Rewards_Receipt_Item_List = pd.DataFrame(Receipt[['_id','rewardsReceiptItemList']])
```

```python
# drop the rewardsReceiptItemList column from the Receipt DataFrame
Receipt.drop(['rewardsReceiptItemList'],axis=1, inplace=True)
```

# Identify Data Quality Issues

```python
# check the data type
data_type = pd.DataFrame(Receipt.dtypes,columns=['Data Type'])

# check for missing data
missing_data = pd.DataFrame(Receipt.isnull().sum(), columns=['Missing Values'])

# check if the values are unique
unique_values = pd.DataFrame(columns=['Unique Values'])
for row in list(Receipt.columns.values):
    unique_values.loc[row] = [Receipt[row].nunique()]

Data_summary = data_type.join(missing_data).join(unique_values)

Data_summary
```

##### There are a lot of missing values, we need to explore why there are missing and how to deal with these missing values
From the summary data above, we see there are 510 missing values of pointsEarned, check why they are missing.

```python
# number of recepits group by reward receipt status
Receipt[['_id','rewardsReceiptStatus']].groupby('rewardsReceiptStatus').count()
```

```python
# number of receipts that the pointsEarned value is missing break by reward receipt status
Receipt[Receipt['pointsEarned'].isnull()][['rewardsReceiptStatus','_id']].groupby('rewardsReceiptStatus').count()
```

##### We find that the number of PENDING and SUBMITTED matches, so the pointsEarned value are missing for those receipts that their status is PENDING or SUBMITTED

```python
# check those receipts that are rejected
Receipt[Receipt['rewardsReceiptStatus'] == 'REJECTED']
```
